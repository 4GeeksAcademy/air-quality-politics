```python
if 'Partido' in df.columns:
    mapa_color = {
        'PSOE': 0, 'PSC': 0, 'UP': 0, 'SUMAR': 0, 'ERC': 0,
        'PP': 1, 'VOX': 1, 'Cs': 1,
        'PNV': 2, 'JxCat': 2, 'BILDU': 2
    }
    df['color_politico'] = df['Partido'].map(mapa_color).fillna(3).astype(int)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Agrupar por año y color político, calcular media de muertes
df_evolucion = df_nuevo.groupby(['year', 'color_politico'])['Muertes_por_causa'].mean().reset_index()

# Mapa de colores y etiquetas para los partidos
etiquetas = {
    0: 'Izquierda',
    1: 'Derecha',
    2: 'Nacionalistas',
    3: 'Otros'
}

# Reemplazar códigos por etiquetas para mejor lectura
df_evolucion['color_label'] = df_evolucion['color_politico'].map(etiquetas)

# Gráfico de líneas
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_evolucion, x='year', y='Muertes_por_causa', hue='color_label', marker='o')
plt.title("Evolución de muertes por causa según color político (2013–2022)")
plt.xlabel("Año")
plt.ylabel("Muertes por causa (media mensual)")
plt.grid(True)
plt.show()
```

```python
# Preparar df_edad_merge desde el archivo el_f_dataset.csv
df_edades = pd.read_csv('/workspaces/air-quality-politics/data/el_f_dataset.csv', delimiter=',', index_col=False)

df_edades = df_edades.rename(columns={
'Anio': 'year',
'Mes': 'month',
'Codigo_Provincia': 'codigo_provincia',
'Grupo9Edad': 'grupo_edad'
})

```

```python
df_edades.tail()
```

```python
df_edades['grupo_edad'].unique()
```

```python
# Definir los grupos considerados como mayores de 65
grupos_mayores_65 = ['65-74', '75+']

# Total población por provincia, año y mes
total_pob = df_edades.groupby(['codigo_provincia', 'year', 'month'])['Total'].sum().reset_index(name='pob_total')

# Filtrar y sumar población mayor de 65
pob_65 = df_edades[df_edades['grupo_edad'].isin(grupos_mayores_65)]
pob_65_sum = pob_65.groupby(['codigo_provincia', 'year', 'month'])['Total'].sum().reset_index(name='pob_65')

# Merge y cálculo del porcentaje
df_edad_merge = pd.merge(total_pob, pob_65_sum, on=['codigo_provincia', 'year', 'month'], how='left')
df_edad_merge['porc_mayores_65'] = (df_edad_merge['pob_65'] / df_edad_merge['pob_total']) * 100
```

```python
df_edad_merge[['codigo_provincia', 'year', 'month', 'porc_mayores_65']].dropna().head()
```

```python
df_nuevo = df_nuevo.merge(
    df_edad_merge[['codigo_provincia', 'year', 'month', 'porc_mayores_65']],
    on=['codigo_provincia', 'year', 'month'],
    how='left'
)
```

```python
df_nuevo = df_nuevo.drop(columns=['porc_mayores_65_x', 'porc_mayores_65_y'], errors='ignore')
```
