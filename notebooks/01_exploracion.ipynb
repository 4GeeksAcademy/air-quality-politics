```python
import glob
import os

archivos = glob.glob("data/processed/*.csv")
ciudades_guardadas = sorted(set(os.path.basename(f).split("_")[0] for f in archivos))
print(f"✅ {len(ciudades_guardadas)} ciudades descargadas:")
print(ciudades_guardadas)
```

```python
from collections import defaultdict

ciudad_años = defaultdict(set)

for f in archivos:
    nombre = os.path.basename(f).replace(".csv", "")
    partes = nombre.split("_")
    if len(partes) == 2:
        ciudad, anio = partes
        ciudad_años[ciudad].add(anio)

# Mostrar ciudades con menos de 11 años (2013–2023)
for ciudad, años in sorted(ciudad_años.items()):
    if len(años) < 11:
        print(f"⚠️ {ciudad}: {sorted(años)} ({len(años)} años)")
```

```python
from extract_aemet import download_aemet
from station_map import station_map

# Lista de combinaciones ciudad-año que faltan
missing = [
    ("Badajoz", 2022),
    ("Barcelona", 2023),
    ("Bilbao", 2017), ("Bilbao", 2018), ("Bilbao", 2019), ("Bilbao", 2020),
    ("Burgos", 2013), ("Burgos", 2014), ("Burgos", 2015), ("Burgos", 2016),
    ("Burgos", 2017), ("Burgos", 2018), ("Burgos", 2019), ("Burgos", 2020),
    ("Castellón de la Plana", 2013), ("Castellón de la Plana", 2014),
    ("Girona", 2013), ("Girona", 2014), ("Girona", 2015), ("Girona", 2016),
    ("Girona", 2017), ("Girona", 2018), ("Girona", 2019), ("Girona", 2020), ("Girona", 2021),
    ("Huesca", 2014), ("Huesca", 2016), ("Huesca", 2017), ("Huesca", 2018), ("Huesca", 2019),
    ("Huesca", 2020), ("Huesca", 2021), ("Huesca", 2022), ("Huesca", 2023),
    ("Santa Cruz de Tenerife", 2013), ("Santa Cruz de Tenerife", 2014), ("Santa Cruz de Tenerife", 2015),
    ("Santa Cruz de Tenerife", 2016), ("Santa Cruz de Tenerife", 2017),
    ("Tarragona", 2013), ("Tarragona", 2014), ("Tarragona", 2015), ("Tarragona", 2016), ("Tarragona", 2023),
    ("Teruel", 2013), ("Teruel", 2016), ("Teruel", 2017), ("Teruel", 2018), ("Teruel", 2019),
    ("Teruel", 2020), ("Teruel", 2021), ("Teruel", 2022), ("Teruel", 2023),
    ("Ávila", 2013), ("Ávila", 2014), ("Ávila", 2019), ("Ávila", 2020), ("Ávila", 2021), ("Ávila", 2022)
]

def download_missing_years():
    for city, year in missing:
        print(f"⬇️ Descargando: {city} - {year}")
        station_id = station_map.get(city)
        if not station_id:
            print(f"⚠️ Ciudad no encontrada en el station_map: {city}")
            continue
        try:
            download_aemet(city, station_id, year)
        except Exception as e:
            print(f"❌ Error en {city}-{year}: {e}")
```
